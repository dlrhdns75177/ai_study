````python
#마크업 : 태그와 텍스트(글자 그 자체)를 이용해서 문서의 구조(제목, 서론, 본론 등으로 나누어 줌)를 나타내는 것 -> 마우스 사용할 필요 x
#마크다운(.md) : 마크업 언어의 일종으로 더 간단하게 표현 가능하다(#, -, ** 등을 사용)
#헤딩 : #의 개수를 가지고 제목의 수준을 구분
#리스트 : 순서가 있는거 없는거 존재
#코드 블럭: ``` ```(여러줄 코드 작성할 떄 )  ` `(한 줄 코드 작성할 떄)
#링크 : [문구](링크 주소 넣기)
#이미지 : ![문구](이미지 경로)
#텍스트 강조 : ** **(안에 문장 넣으면 두꺼워짐) * *(안에 문장 넣으면 기울임체) ~~ ~~(안에 적으면 취소)
#수평선 : ---으로 경계선 그리기
#인용문 : >
# 표 : |이름|나이|학년| 이런식으로

#깃 : 분산(중앙 집중의 반대 의미로 파일 하나를 그대로 두고 각 변경사항만 기억한다) 버전 관리 시스템(코드의 히스토리 관리(전과 후))
#깃허브 : 깃을 저장하기 위해서 만들어진 저장소(서비스)
# ls :현재 위치의 폴더, 파일 목록 보기
# ls -al :현재 위치의 숨김 파일을 포함하여 보기
# ~ 경로 : user이름의 가장 상위 폴더
# pwd : 현재 위치 보기
# cd path: path로 현재 위치에서 하위로 이동하기
# cd .. : 상위 폴더로 이동
# mkdir name :name으로 폴더 생성하기
# touch name :name으로 파일 생성하기
# rm name :파일 삭제하기
# rm -r name :폴더 삭제하기(한글이나 띄어쓰기가 있으면 ' ' 로 묶어주기)
#repository : 특정 디렉토리(폴더)를 버전 관리하는 저장소(버전관리를 시작하는 폴더부터 버전 관리 즉 그 하위 폴더들은 다 버전관리 하지만 상위는 버전관리 ㄴㄴ)
# 숨김 파일 : .으로 시작하는 파일들
#master : 깃으러 저정되고 있다
# git init : 버전 관리를 하고 싶은 파일에 들어가서 
# 로컬 저장소를 생성(.git 파일이 생김)
# .git :이 디렉토리 안에는 버전 관리에 필요한 모든 것이 들어 있다(아직 버전관리 시작 x)
````
