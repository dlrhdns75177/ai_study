```python
import pandas as pd #판다스 라이브러리 불러오기
df1 = pd.read_excel('crime.xlsx') #액셀 파일 불러오기
df1
```

```python
mapping = { #매핑할 딕셔너리 만들기
    '서대문서': '서대문구', '수서서': '강남구', '강서서': '강서구', '서초서': '서초구',
'서부서': '은평구', '중부서': '중구', '종로서': '종로구', '남대문서': '중구',
'혜화서': '종로구', '용산서': '용산구', '성북서': '성북구', '동대문서': '동대문구',
'마포서': '마포구', '영등포서': '영등포구', '성동서': '성동구', '동작서': '동작구',
'광진서': '광진구', '강북서': '강북구', '금천서': '금천구', '중랑서': '중랑구',
'강남서': '강남구', '관악서': '관악구', '강동서': '강동구', '종암서': '성북구', 
'구로서': '구로구', '양천서': '양천구', '송파서': '송파구', '노원서': '노원구', 
'방배서': '서초구', '은평서': '은평구', '도봉서': '도봉구'
}

df1['구별'] = df1['관서명'].map(mapping).fillna('구 없음') #구별이라는 새로운 열에 기존 데이터프레임의 관서명을 map하고 결측값은 '구 없음'으로 채우기
df1
```

```python
pivot_df1 = pd.pivot_table(df1, index='구별',aggfunc='sum') #피벗 테이블로 구별을 기준으로 합 구하기
pivot_df1
```

```python
df1 = df1.drop(df1[df1['구별']=='구 없음'].index) #데이터프레임에서 구별이라는 열 중에서 '구 없음'행은 삭제
df1
```

```python
list = ['강간','강도','살인','절도','폭력','소계'] #리스트 만들기
for i in list: #반복문으로 모든 리스트 실행
    if i == '소계': #소계일 경우에는
        df1['검거율'] = df1[i+'(검거)']/df1[i+'(발생)']*100 #이렇게 계산
    else: #그 외에는
        df1[i+'검거율'] = df1[i+'(검거)']/df1[i+'(발생)']*100 #이렇게 계산
df1
```

```python
del_list = ['강간','강도','살인','절도','폭력','소계'] #삭제할 리스트 만들기
for i in del_list: #반복문 실행
    if i == '소계': #만약에 소계일떄는
        del df1[i+'(검거)'] #이렇게 삭제
        del df1[i+'(발생)'] #이렇게 삭제
    else: #그 외에는 
        del df1[i+'(검거)'] #이렇게 삭제
df1
```

```python
name = { #새롭게 바꿀 이름을 딕셔너리 형태로 만들기
    '강간(발생)':'강간',
    '강도(발생)':'강도',
    '살인(발생)':'살인',
    '절도(발생)':'절도',
    '폭력(발생)':'폭력'
}
df1.rename(columns=name, inplace=True) #rename함수로 이름 바꾸기
df1.set_index('구별', inplace=True) #set_index함수로 구별 열을 인덱스로 설정
del df1['관서명'] #관서명 열 삭제
df1
```

```python
df2 = pd.read_csv('pop_kor.csv').set_index('구별') #csv 파일 불러오기 그리고 인덱스를 구별로 설정
df2
```

```python
df3 = df1.join(df2) #join함수로 df1이랑 df2 인덱스를 기준으로 합치기
df3
```

```python
df3.sort_values(by='검거율', ascending=True) #sort_values 함수를 이용해서 검거율 기준으로 오름차순 정렬
```
